```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/rajsharma/Downloads/netflix_titles 2.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['director'] = data['director'].fillna("UnKnown")
data['cast'] = data['cast'].fillna("UnKnown")
data['country'] = data['country'].fillna("UnKnown")
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=['date_added', 'rating', 'duration'])
```

```python
data.isnull().sum()
```

```python
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
```

```python
data.head()
```

```python
data[pd.isna(data['date_added'])]
```

```python
data['date_added'].isna().sum()
```

```python
data = data.dropna(subset=['date_added'])
```

```python
data['date_added'].isna().sum()
```

```python
data['duration_int'] = data['duration'].str.extract('(\d+)').astype(float)

data.loc[data['type'] == 'Movie' , 'duration_type'] = 'minutes'
data.loc[data['type'] == 'TV Show' , 'duration_type'] = 'seasons'
```

```python
data.head()
```

```python
data = data.drop(columns=['duration'])
```

```python
data.head()
```

```python
data['main_country'] = data['country'].str.split(',').str[0]

data['genre'] = data['listed_in'].str.split(',').str[0]

data.head()
```

```python
counts = data['type'].value_counts()
counts
```

```python
plt.figure(figsize = (10,6))
plt.bar(counts.index , counts.values , color = ['skyblue' , '#808000'])

for i, val in enumerate(counts.values):
    plt.text(i,val/2, str(val), ha='center',va = 'center', fontsize=12)

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type of Content")
plt.ylabel("Number of Titles")
plt.show()   
```

```python

data['release_year'] = pd.to_numeric(data['release_year'], errors='coerce')
growth = data.groupby("release_year").size()
plt.figure(figsize=(10,5))
plt.plot(growth.index, growth.values, marker='o')
plt.title("Netflix Content Growth Over Time")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles Released")
plt.grid(True, linestyle="-", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
growth_split = data.groupby(['release_year','type']).size().unstack(fill_value=0)

plt.figure(figsize=(10,5))
plt.plot(growth_split.index, growth_split['Movie'], label="Movies", marker='o')
plt.plot(growth_split.index, growth_split['TV Show'], label="TV Shows", marker='s')
plt.title("Netflix Content Growth Over Time (Movies vs TV Shows)")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles Released")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
data['added_year'] = data['date_added'].dt.year
data.head()
```

```python
added = data.groupby(['added_year' , 'type']).size().unstack(fill_value = 0)

plt.figure(figsize= (10,6))
plt.plot(added.index , added['Movie'] , label = 'Movies' , marker = 'o')
plt.plot(added.index , added['TV Show'] , label = 'Tv Shows', marker = 'x')

plt.title('Content Added trough years(Movies Vs Tv Shows)')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.legend()
plt.grid(True , linestyle = '-' , alpha = 0.4)
#plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
country_count = data[data['country'] != "Unknown"]['country'].str.split(',').explode().str.strip().value_counts()
top_country = country_count.head(10)

plt.figure(figsize=(12, 6))
top_country.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Split country values and explode

#df_expanded = data.dropna().copy()
df_expanded['country'] = data['country'].str.split(',').explode('country').str.strip()

# Group by country and type
country_type = df_expanded.groupby(['country','type']).size().unstack(fill_value=0)

# Top 10 countries by total content
top_countries_split = country_type.sum(axis=1).sort_values(ascending=False).head(10)
country_type_top10 = country_type.loc[top_countries_split.index]

# Stacked bar chart
country_type_top10.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title("Top 10 Countries Producing Netflix Content (Movies vs TV Shows)")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python


genere_count = data['listed_in'].str.split(',').explode().str.strip().value_counts()

top_genre = genere_count.head(10)

plt.figure(figsize = (10,6))
top_genre.plot(kind = 'bar' , color = 'skyblue' , edgecolor = 'black')

for i, val in enumerate(top_genre.values):
    plt.text(i,val+50, str(val), ha='center',va = 'center', fontsize=12)
plt.title("top genre of Movies and TvShows")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
genre_count = data.copy()

genre_count['listed_in'] = data['listed_in'].str.split(',').explode('listed_in').str.strip()

genre_type = genre_count.groupby(['listed_in','type']).size().unstack(fill_value = 0)

top_genre_split = genre_type.sum(axis  = 1).sort_values(ascending = False).head(10)

genre_top10 = genre_type.loc[top_genre_split.index]

genre_top10.plot(kind = 'bar' , stacked = True,figsize=(12,6))
plt.title("Top 10 Genre Producing Netflix Content (Movies vs TV Shows)")
plt.xlabel("Genre")
plt.ylabel("Numbers of Genre")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
rating_counts = data['rating'].dropna().value_counts().head(10)  # Top 10 ratings

plt.figure(figsize=(10,6))
rating_counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Distribution of Ratings (Top 10)")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
movies = data[data['type'] == 'Movie']

# Keep only rows where duration_type is 'minutes'
movies = movies[movies['duration_type'] == 'minutes']

# Calculate average duration
avg_duration = movies['duration_int'].mean()
print(f"📽️ Average Movie Duration: {avg_duration:.2f} minutes")

plt.figure(figsize=(10,6))
plt.hist(movies['duration_int'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Movie Durations")
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
```
